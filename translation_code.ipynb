```python
import pandas as pd
import googletrans
from googletrans import *
translator=googletrans.Translator()
```

```python
data_t=pd.read_csv("/content/tbilisi_emoji.csv", sep=',',  lineterminator='\n')
```

```python
data_t
```

```python
data_t["reviews"]=data_t["reviews"].astype(str)
data_t["translated"]=data_t["reviews"].apply(translator.translate, src='auto', dest='en').apply(getattr, args=('text',))
```

```python
data_t
```

```python
data_t.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_t=data_t
```

```python
df_t
```

```python
df_t.to_csv("tbilisi_translated.csv")
```
